```python
# Audit checks
def audit_file(name, file):
    df = pd.read_csv(file, low_memory=False)

    if "2011" in name:
        year = "2011"
    else:
        year = "2021"

    if "msoa" in name:
        geography = "MSOA"
    else:
        geography = "LA"

    if "age" in name:
        theme = "age"
    elif "sex" in name:
        theme = "sex"
    elif "ethnicity" in name:
        theme = "ethnicity"
    elif "unpaid_care" in name:
        theme = "unpaid care"
    elif "economic_activity" in name:
        theme = "economic activity"
    else:
        theme = "unknown"

    code_col = "geography code" if "geography code" in df.columns else "not found"
    name_col = "geography" if "geography" in df.columns else "not found"

    if code_col != "not found":
        duplicate_codes = df[code_col].duplicated().sum()
        unique_codes = df[code_col].nunique()
    else:
        duplicate_codes = None
        unique_codes = None

    value_cols = df.select_dtypes(include="number").columns
    negative_values = int((df[value_cols] < 0).sum().sum()) if len(value_cols) > 0 else 0

    return {
        "file": file.name,
        "year": year,
        "geography": geography,
        "theme": theme,
        "rows": len(df),
        "columns": len(df.columns),
        "missing_values": int(df.isna().sum().sum()),
        "duplicate_codes": duplicate_codes,
        "unique_codes": unique_codes,
        "negative_values": negative_values,
        "key_join_column": code_col,
        "area_name_column": name_col,
        "first_few_columns": "; ".join(list(df.columns[:6])),
    }

audit_rows = []

for name, file in files.items():
    audit_rows.append(audit_file(name, file))

audit_table = pd.DataFrame(audit_rows)
display(audit_table)
```

```python
"""
Check geography counts:  2011 and 2021 Local Authority files do not have the same number of areas
"""

la_2011_codes = pd.read_csv(files["2011_age_la"])["geography code"]
la_2021_codes = pd.read_csv(files["2021_age_la"])["geography code"]

matched_codes = set(la_2011_codes).intersection(set(la_2021_codes))
only_2011 = sorted(set(la_2011_codes) - set(la_2021_codes))
only_2021 = sorted(set(la_2021_codes) - set(la_2011_codes))

print("2011 LA codes:", la_2011_codes.nunique())
print("2021 LA codes:", la_2021_codes.nunique())
print("matched codes:", len(matched_codes))
print("2011-only codes:", len(only_2011))
print("2021-only codes:", len(only_2021))
```

```python
"""
The lookup files are needed for two reasons:

1. To convert 2011 Local Authorities onto 2021 boundaries
2.To connect 2021 MSOAs to their Local Authority

"""

msoa_lookup = pd.read_csv(lookup_files["msoa_to_la"], dtype=str, low_memory=False)
la_lookup = pd.read_csv(lookup_files["la_2011_to_2021"], dtype=str)

print("msoa lookup shape:", msoa_lookup.shape)
print("la lookup shape:", la_lookup.shape)

print("\nmsoa lookup columns:")
print(list(msoa_lookup.columns))

print("\nla lookup columns:")
print(list(la_lookup.columns))
```

```python
msoa_to_la = msoa_lookup[["MSOA21CD", "MSOA21NM", "LAD22CD", "LAD22NM"]].drop_duplicates()
la_2011_to_2021 = la_lookup[["LAD11CD", "LAD11NM", "LAD21CD", "LAD21NM"]].drop_duplicates()

print("unique msoas in lookup:", msoa_to_la["MSOA21CD"].nunique())
print("unique local authorities in msoa lookup:", msoa_to_la["LAD22CD"].nunique())

check = msoa_to_la.groupby("MSOA21CD")["LAD22CD"].nunique()
print("msoas linked to more than one la:", (check > 1).sum())

print("\n2011 authorities in lookup:", la_2011_to_2021["LAD11CD"].nunique())
print("2021 authorities in lookup:", la_2011_to_2021["LAD21CD"].nunique())
```

```python
# Unpaid care data
age_total_2011 = "Age: All categories: Age; measures: Value"
care_total_2011 = "Care Provision: All categories: Provision of unpaid care; measures: Value"

age_2011 = pd.read_csv(files["2011_age_la"])[["geography code", age_total_2011]]
care_2011 = pd.read_csv(files["2011_unpaid_care_la"])[["geography code", care_total_2011]]

care_check_2011 = age_2011.merge(care_2011, on="geography code")
care_check_2011["difference"] = care_check_2011[age_total_2011] - care_check_2011[care_total_2011]

print("2011 max absolute difference between age total and care total:")
print(care_check_2011["difference"].abs().max())
```

```python
age_total_2021 = "Age: Total; measures: Value"
care_total_2021 = "Provision of unpaid care: Total: All usual residents aged 5 and over"

age_2021 = pd.read_csv(files["2021_age_la"])[["geography code", age_total_2021]]
care_2021 = pd.read_csv(files["2021_unpaid_care_la"])[["geography code", care_total_2021]]

care_check_2021 = age_2021.merge(care_2021, on="geography code")
care_check_2021["age_minus_care_denominator"] = care_check_2021[age_total_2021] - care_check_2021[care_total_2021]

print("2021 max difference between age total and care denominator:")
print(care_check_2021["age_minus_care_denominator"].max())

display(care_check_2021.sort_values("age_minus_care_denominator", ascending=False).head())
```

```python
# checking join works well

msoa_codes = pd.read_csv(files["2021_age_msoa"])[["geography code"]].rename(columns={"geography code": "msoa_code"})

msoa_to_la_clean = msoa_to_la.rename(columns={
    "MSOA21CD": "msoa_code",
    "MSOA21NM": "msoa_name",
    "LAD22CD": "la_code_2021",
    "LAD22NM": "la_name_2021"
})

joined = msoa_codes.merge(msoa_to_la_clean, on="msoa_code", how="left")

print("msoa rows:", len(joined))
print("missing local authority codes after join:", joined["la_code_2021"].isna().sum())
print("unique local authorities after join:", joined["la_code_2021"].nunique())
```
